## Vectorizing across m examples

### Crude Format (without vectorization)

for i = 1 to m:

$$ Z^{[1](i)} = W^{[1]}x^{(i)} + b ^{[1]}$$
$$ a^{[1](i)} = \sigma(Z^{[1](i)})$$
$$ Z^{[2](i)} = W^{[2]}a^{[1](i)} + b^{[2]} $$
$$ a^{[2](i)} = \sigma(Z^{[2](i)}) $$

Note: [l] -> says layer number, 
(i) -> denotes example number

### Vectorizing the above equation

Let us stack all input features onto a matrix and it's given as follows:

$$ X = \begin{bmatrix} 
x_{1}^{(1)} & .. & x_{1}^{(m)} \\
x_{2}^{(1)} & .. & x_{2}^{(m)} \\
.. & .. & ..\\
x_{n_{x}}^{(1)} & .. & x_{n_{x}}^{(m)} \\
\end{bmatrix}$$


$$ Z^{[1]} = \begin{bmatrix} 
. & .. & . \\
Z^{[1](1)} & .. & Z^{[1](m)} \\
. & .. & .\\
\end{bmatrix} $$

$$ A^{[1]} = \begin{bmatrix} 
. & .. & . \\
a^{[1](1)} & .. & a^{[1](m)} \\
. & .. & .\\
\end{bmatrix} $$


The vectorized form is given below

$$ Z^{[1]} = W^{[1]}X + b ^{[1]}$$
$$ A^{[1]} = \sigma(Z^{[1]})$$
$$ Z^{[2]} = W^{[2]}A^{[1]} + b^{[2]} $$
$$ A^{[2]} = \sigma(Z^{[2]}) $$

In the above equations,
size of $X = (n_{x},m)$
$W^{[1]} = (k, n_{x})$
where k is the number of neurons


## Activation functions

Different layer can have different activation functions.

For instance, 

1. Sigmoid function: Range: 0 < a < 1
$$ a = \sigma(z) = \frac{1}{1+e^{-z}}$$

2. Hyperbolic tangent: Range: -1 < a < 1
$$ a = \tanh(z) = \frac{e^{z}-e^{-z}}{e^{z}+e^{-z}} $$

Downsides of above two functions, when z is very large or small, the gradient becomes very small ( slope -> 0)and makes it difficult to optimize.

3. Rectified Linear Unit: 

$$a=max(0,z)$$
Derivative: z > 0, slope = 1 
z < 0, slope = -1


Rules: 
1. For Binary classification, sigmoid activation function (for output layer), for all other units (or hidden layers) use RelU.
